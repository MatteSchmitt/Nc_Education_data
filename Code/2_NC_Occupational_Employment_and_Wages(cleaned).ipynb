```python
import pandas as pd
import numpy as np
```

```python
file_path = '2_NC_Occupational_Employment_and_Wages.csv'
df2 = pd.read_csv(file_path)
df2.head()
```

```python
df2.replace('*', np.nan, inplace=True)
df2.head()
```

```python
df2.to_csv('2_NC_Occupational_Employment_and_Wages(cleaned)csv', index=False)
```

```python
if df2['Annual wage; mean'].isnull().any():
    print("There are missing values in the 'Annual wage; mean' column.")
else:
    print("No missing values in the 'Annual wage; mean' column.")
```
